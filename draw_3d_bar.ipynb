```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# 
fig = plt.figure()
ax = plt.axes(projection='3d')

# 
x = np.arange(6)  # x56
y = np.arange(5)  # y65
X, Y = np.meshgrid(x, y)  # 
print("X and Y:")
print(X)
print(Y)
Z = np.random.rand(5, 6)
print("Z:")
print(Z)
print()
print("X.flatten:")
print(X.flatten())
print("Y.flatten:")
print(Y.flatten())

print("np.zeros_like(Z): ")
print(np.zeros_like(Z))

print("np.zeros_like(Z).flatten(): ")
print(np.zeros_like(Z).flatten())
# 
ax.bar3d(X.flatten(), Y.flatten(), np.zeros_like(Z).flatten(), 0.5, 0.5, Z.flatten(), color='b')

# 
ax.set_xlabel('X ')
ax.set_ylabel('Y ')
ax.set_zlabel('Z ')

# 
plt.show()
```
