```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# 创建一个三维坐标系
fig = plt.figure()
ax = plt.axes(projection='3d')

# 生成随机数据
x = np.arange(6)  # 将x数组的长度由5改为6
y = np.arange(5)  # 将y数组的长度由6改为5
X, Y = np.meshgrid(x, y)  # 生成网格数据
print("X and Y:")
print(X)
print(Y)
Z = np.random.rand(5, 6)
print("Z:")
print(Z)
print()
print("X.flatten:")
print(X.flatten())
print("Y.flatten:")
print(Y.flatten())

print("np.zeros_like(Z): ")
print(np.zeros_like(Z))

print("np.zeros_like(Z).flatten(): ")
print(np.zeros_like(Z).flatten())
# 绘制三维柱状图
ax.bar3d(X.flatten(), Y.flatten(), np.zeros_like(Z).flatten(), 0.5, 0.5, Z.flatten(), color='b')

# 设置坐标轴标签
ax.set_xlabel('X 轴')
ax.set_ylabel('Y 轴')
ax.set_zlabel('Z 轴')

# 显示图形
plt.show()
```
